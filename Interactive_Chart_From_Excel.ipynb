## This code does the following:

Loads the data from the "Data" sheet of the "Financial_Data.xlsx" workbook.

Checks for the required columns ("Country" and "Sales").

Groups the data by country and calculates the total sales for each country.

Creates an interactive bar chart using plotly.express.

Saves the chart as an HTML file in the same directory as the workbook.

Displays the chart using fig.show()

## Code 1: Using plotly.express and plotly.graph_objects

```python
import pandas as pd
import plotly.express as px
import os

file_path = '/content/interactive-chart-excel/Financial_Data.xlsx'

try:
    # load the data from the "Data" sheet of the "Financial_Data.xlsx" workbook
    data = pd.read_excel(file_path, sheet_name="Data")

    # check if 'Country' and 'Sales' columns exists in data
    if 'Country' not in data.columns or 'Sales' not in data.columns:
        raise ValueError("Columns are missing")
    # group the data by country and calculate the total sales for each country
    sales_by_country = data.groupby("Country")["Sales"].sum().reset_index()

    # create the bar chart
    fig = px.bar(sales_by_country, x="Country", y="Sales", title="Financial Data By Country",
                 labels={"Country": "Country", "Sales": "Total Sales"},
                 color_discrete_sequence=["#00008B"])

    # save the chart to the same directory as the workbook
    fig.write_html("Financial Data By Country.html")

    # display the chart
    fig.show()

except FileNotFoundError as e:
    print(f"{e} not found")
except ValueError as e:
    print(e)
```

## Code 2: Using matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/interactive-chart-excel/Financial_Data.xlsx'

try:
    # Load the data from the "Data" sheet of the "Financial_Data.xlsx" workbook
    data = pd.read_excel(file_path, sheet_name="Data")

    # Check if 'Country' and 'Sales' columns exist in data
    if 'Country' not in data.columns or 'Sales' not in data.columns:
        raise ValueError("Columns are missing")

    # Group the data by country and calculate the total sales for each country
    sales_by_country = data.groupby("Country")["Sales"].sum().reset_index()

    # Create the bar chart
    plt.figure(figsize=(10, 6))
    plt.bar(sales_by_country["Country"], sales_by_country["Sales"], color="#00008B")
    plt.xlabel("Country")
    plt.ylabel("Total Sales")
    plt.title("Financial Data By Country")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

except FileNotFoundError as e:
    print(f"{e} not found")
except ValueError as e:
    print(e)
```
